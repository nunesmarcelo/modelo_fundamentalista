```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import zipfile
```

```python
class Manipulation():
    def __init__(self):
        self.frames_list = {}
        print("Instantiated class!")
        
    def ler_dados_empresa(self, file):
        zf = zipfile.ZipFile(file)
        empresa = pd.read_excel(zf.open('balanco.xls') , encoding='ISO-8859-1', sheet_name="Bal. Patrim.")
        nome_empresa = empresa.columns[1].split('-')[1].strip().lower()
        
        dados_historicos = pd.read_excel(zf.open('balanco.xls') , encoding='ISO-8859-1', skiprows=1, sheet_name="Bal. Patrim.")

        dados_historicos = dados_historicos.transpose()
        dados_historicos.columns = dados_historicos.iloc[0]
        dados_historicos = dados_historicos.drop('Unnamed: 0')
        dados_historicos.columns.name = "Data"
        dados_historicos.index = pd.to_datetime(dados_historicos.index)
        
        dados_dre = pd.read_excel(zf.open('balanco.xls') , encoding='ISO-8859-1', skiprows=1, sheet_name="Dem. Result.")
        dados_dre = dados_dre.transpose()
        dados_dre.columns = dados_dre.iloc[0]
        dados_dre = dados_dre.drop(dados_dre.index[0])
        dados_dre.columns.name = "Data"
        dados_dre.index = pd.to_datetime(dados_dre.index)
        
        total = dados_historicos.reset_index().merge(dados_dre.reset_index(), on="index")
        total = total.set_index("index")
        
        self.frames_list[nome_empresa] = total
        
    def imprimir(self, enterprise):
        empresas = list(self.frames_list.keys())
        
        
        if enterprise in empresas:
            display(self.frames_list[enterprise])
        else:
            print("Empresas disponíveis:" , empresas)
        
```

```python
arquivo = "bal_ITA.zip"
cls = Manipulation()
cls.ler_dados_empresa(arquivo)
cls.imprimir("itaúsa")
```
